```python
import numpy as np
import pandas as pd
import math
from pandas import ExcelWriter
from pandas import ExcelFile
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn import linear_model as lm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
data = pd.read_csv("df2.csv")
```

```python
features = pd.read_csv('train.csv', usecols=['TotalBsmtSF','1stFlrSF','YearRemodAdd','GarageArea'])
saleprice=data['SalePrice']
baseline_saleprice= saleprice.mean()
```

```python
# Linear Regression Model
```

```python
print("Test 1")
X_train, X_test, y_train, y_test = train_test_split(features, saleprice, test_size = 0.20, random_state = 640)

model = lm.LinearRegression()

model = model.fit(X=X_train, y=y_train)
predictions = model.predict(X=X_test)

mse = sklearn.metrics.mean_squared_error(y_true = y_test, y_pred = predictions)
rmse = math.sqrt(mse)
print("rmse: " + str(rmse))

baseline = [y_test.mean()]*len(y_test)
mse2 = sklearn.metrics.mean_squared_error(y_true = y_test, y_pred = baseline)
rmse2 = math.sqrt(mse2)
accuracy = (rmse2 - rmse)/(rmse2) * 100
print("baseline: " + str(rmse2))
print("accuracy: " + str(accuracy))

plt.figure()
plt.scatter(X_test['GarageArea'],y_test)
plt.scatter(X_test['YearRemodAdd'],y_test)
plt.scatter(X_test['TotalBsmtSF'],y_test)
plt.scatter(X_test['1stFlrSF'],y_test)
plt.title('Housing Price Metrics')
plt.show()
```

```python
features2 = pd.read_csv('df2.csv', usecols=['OverallCond','YearRemodAdd'])
buildyear=data['YearBuilt']
```

```python
print('Split 1')
X_train, X_test, y_train, y_test = train_test_split(features2, buildyear, test_size = 0.20, random_state = 61)
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train) 
predict = knn.predict(X_test)


actual = np.zeros(len(X_test))
prediction = np.zeros(len(X_test))

zeroes = 0
ones = 0
y_test.index=range(len(X_test))
for x in range(len(X_test)):
    if(y_test[x]< 1970):
        actual[x] = 0
        zeroes += 1
    elif(y_test[x] >= 1970):
        actual[x] = 1
        ones += 1
        
baseline = 0

if(ones >= zeroes):
    baseline = ones/(ones+zeroes)
else:
    baseline = zeroes/(ones+zeroes)
    

for x in range(len(X_test)):
    if(predict[x]< 1970):
        prediction[x] = 0

    elif(predict[x] >= 1970):
        prediction[x] = 1

        
predictioncount = 0
for x in range(len(X_test)):
    if(prediction[x]==actual[x]):
        predictioncount += 1
        
predictionvalue= predictioncount/(ones+zeroes)
accuracy = (predictionvalue-baseline)/baseline

print("baseline: " + str(baseline))
print("accuracy: " + str(accuracy))
```
