```python
import pandas as pd
housing_data=pd.read_csv("train.csv")
housing_data
housing_data.describe()
df = pd.DataFrame(housing_data)
```

```python
y = []
for x in df.columns:
    if df[x].isnull().values.any() and df[x].dtype == "float64":
        y = y + [x]
        
for k in y:
    df[k] = df[k].fillna(value = df[k].mean())
    
df    
```

```python
z=[]
for x in df.columns:
    if df[x].isnull().values.any() and df[x].dtype == "object":
        z = z + [x]
        
for k in z:
    df[k] = df[k].fillna(value = "none")
    
df    
    
```

```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt

import seaborn as sns
fig4 = plt.figure()

plt.subplot()

# A plot with REALLY high bandwith
sns.kdeplot(df['LotArea'], bw=1460)

plt.title ('Lot Area Density Plot: Bandwidth = 10')
plt.xlabel('Lot Area Width Value')
plt.ylabel('Density')

plt.show()

```

```python
fig3 = plt.figure()
plt.subplot()

plt.hist(df['LotArea'],bins=100)
plt.title('Lot Area Histogram')
plt.xlabel('Lot Area Width')
plt.ylabel('Frequency')
plt.show()
```

```python
fig = plt.figure()
plt.scatter(df["YrSold"],df["SalePrice"])  
plt.title('Year Sold vs. Sale Price")
plt.xlabel('Year Sold')
plt.ylabel('Sale Price')
plt.show(fig)
```

```python
#Gaussian heatmap 2
fig5 = plt.figure()
Z, hmx, hmy = np.histogram2d(df['YrSold'],df['SalePrice'])
plt.title('Year Sold vs. Sale Price')
plt.xlabel('Year Sold')
plt.ylabel('Sale Price')

plt.pcolormesh(hmx,hmy,Z.T)
plt.savefig('gaussian_heatmap.png',bbox_inches='tight')
plt.show()
```
